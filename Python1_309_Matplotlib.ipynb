```python
"""
Интересные графики в стиле xkcd: 
http://nbviewer.ipython.org/url/jakevpd.github.com/downloads/notebooks/XKCD_plots.ipynb
не работает!
"""


from pylab import *
```

```python
x = linspace(0, 5, 10)
y = x ** 2
print(x)
print(y)
```

```python
figure()
plot(x, y, "r")
xlabel("x")
ylabel("y")
title("Title")
show()
```

```python
%matplotlib inline #рисуем прямо в окне браузера
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #левый, нижний, толщина, высота

axes.plot(x,y, "r")
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("Title")
show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #основные оси
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) #дополнительные оси на врезке

#основной график
axes1.plot(x, y, "r") #"r" - это красный
axes1.set_xlabel("x")
axes1.set_ylabel("y")
axes1.set_title("Title")

#врезка
axes2.plot(y, x, "g") #"g" - это зеленый
axes2.set_xlabel("y")
axes2.set_ylabel("x")
axes2.set_title("Insert title");
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

for ax in axes:
    ax.plot(x, y, "r")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_title("title")
    
fig.tight_layout()
```

```python
#можем настраивать размер фигуры

fig, axes = plt.subplots(figsize=(12, 3)) #размер в дюймах

axes.plot(x,y, "r")
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("Title");
```

```python
#рисуем несколько графиков на одном
fig, ax = plt.subplots()

ax.plot(x, x**2, label="y = x**2")
ax.plot(x, x**3, label="y = x**3")
ax.legend(loc=2); #верхний левый угол
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("title");
```

```python
#построение гистограмм

from numpy import *
n = random.randn(100000) #генерируем 100000 случайных чисел 

fig, axes = plt.subplots(1, 2, figsize=(12,4))

axes[0].hist(n, bins=25)
axes[0].set_title("Normal histogram")
axes[0].set_xlim(min(n), max(n))

axes[1].hist(n, cumulative=True, bins=50)
axes[1].set_title("Cumulative detailed histogram")
axes[1].set_xlim((min(n), max(n)));
```
