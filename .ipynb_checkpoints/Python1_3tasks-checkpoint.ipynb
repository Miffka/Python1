```python
#задача 1 - список игр с результатом матча, на выходе - сводная таблица результатов всех матчей
#победа - 3 очка, поражение - 0, ничья - 1
#вывод - Команда:Всего_игр Побед Ничьих Поражений Всего_очков

task11 = [] #инициализируем массив
task12 = {} #инициализируем словарь
n = int(input()) #количество завершенных игр
for i in range(n):
    task11.append([j for j in input().split(";")]) #считываем данные в массив
    
    """
    В этом блоке команд мы составим словарь в виде Название_команды: [Игр, Побед(=0), Ничьих(=0), Поражений(=0), Очков(=0)]
    """
    for j in range(len(task11[i])//2): 
        if task11[i][j*2] in task12: #если название команды уже есть в словаре
            #print("yes")
            task12[task11[i][j*2]][0] += 1 #увеличиваем количество игр на одну
            
        else: #если названия команды еще нет в словаре
            #print("no")
            task12[task11[i][j*2]] = [1] #задаем начальное количество игр
            
            task12[task11[i][j*2]] += [0] #для начала задаем количество побед 0
            task12[task11[i][j*2]] += [0] #и количество ничьих - тоже 0
            task12[task11[i][j*2]] += [0] #и количество поражений - 0
            
            task12[task11[i][j*2]] += [0] #добавляем новый элемент в список -  количество набранных очков, пока равно 0

"""
В этом блоке команд мы заполним в словаре поля Побед, Ничьих, Поражений и Очков
"""
for i in range(n): #пробегаем по всем значениям от 0 до n-1
    if int(task11[i][1]) > int(task11[i][3]): #если количество очков, набранных командой1, больше, чем у команды2
        task12[task11[i][0]][1] += 1 #прибавляем в поле "Победы" команды1 единицу
        task12[task11[i][2]][3] += 1 #и в поле "Поражения" команды2 - единицу
        task12[task11[i][0]][4] += 3 #и в поле "Очки" команды1 - 3 очка за победу    
        
    elif int(task11[i][1]) < int(task11[i][3]): #если количество очков, набранных командой1, меньше, чем у команды2
        task12[task11[i][0]][3] += 1 #прибавляем в поле "Поражения" команды1 единицу
        task12[task11[i][2]][1] += 1 #и в поле "Победы" команды2 - единицу
        task12[task11[i][2]][4] += 3 #и в поле "Очки" команды2 - 3 очка за победу
        
    elif int(task11[i][1]) == int(task11[i][3]): #если команды набрали равное количество очков
        task12[task11[i][0]][2] += 1 #в поле "Ничьи" команды1 добавляем единицу
        task12[task11[i][2]][2] += 1 #и в поле "Ничьи" команды2 тоже добавляем единицу
        task12[task11[i][0]][4] += 1 #и в поле "Очки" команды1 - 1 очко за ничью
        task12[task11[i][2]][4] += 1 #и в поле "Очки команды2 - 1 очко за ничью

#print(task11)
#print(task12)

"""
В этом блоке команд мы выведем наш словарь в требуемом виде на стандартный вывод
"""
for key, value in task12.items(): #пробегаем все ключи и значения в словаре
    print(key + ":", end = "") #печатаем название команды и двоеточие, остаемся на этой строке
    
    for i in value: #пробегаем по всем полям в списке значений
        print(i, end = " ") #печатаем значения в строку
    print() #переходим на другую строку
```

```python
#Задача 2 - подстановочный шифр
"""
Описание кода - первая строка - ключи, вторая строка - значения, третья строка - прямая шифровка, четвертая строка - расшифровка
"""
task21 = [] #инициализируем массив для чтения ввода
task22 = {} #инициализируем словарь для хранения шифра

"""
В этом блоке мы читаем все введенные строки в наш массив
"""
for i in range(4):
    task21.append(list(input()))
#print(task21)

"""
В этом блоке мы записываем шифр в словарь, где ключи - это первая введенная строка, а значения - вторая введенная строка
"""
for i in range(len(task21[0])):
    task22[task21[0][i]] = task21[1][i]
#print(task22)

"""
В этом блоке мы выполним задачу шифровки третьей строки и выведем результат на стандартный вывод
"""
for i in task21[2]: #перебираем все буквы из третьей строки
    print(task22[i], end = "") #печатаем значения согласно словарю
print() #перейдем на следующую строку

"""
В этом блоке мы выполним задачу дешифровки четвертой строки и выведем результат на стандартный вывод
"""
for i in task21[3]: #перебираем все буквы из четвертой строки
    for key, value in task22.items(): #в массиве из ключей из значений словаря
        if i == value: #если буква из четвертой строки соответствует значению
            print(key, end = "") #печатаем ключи в одну строку
            
```

```python
#Задача 3 - система проверки орфографии
"""
Каждое слово в проверяемом тексте ищется в списке известных слов, если слово не найдено, оно помечается как ошибочное.
Первая строка ввода - количество d записей в списке известных слов.
Следующие d строк - сами эти известные слова.

Следующая строка - количество строк на проверку (L строк).
Следующие L строк - сами строки на проверку.

Задача - передать на вывод в новых строках слова из строк на проверку, которые не встречаются в списке из известных строк.
Примечание - регистр не учитывается (ага, все приводим к нижнему).
"""

"""
В этом блоке мы запишем все известные слова в список
"""
task31 = [] #в этот список мы запишем все известные слова

d = int(input()) #задаем количество слов в списке известных слов


for i in range(d): #пробегаем в цикле все значения от 0 до d-1
    task31.append(input().lower()) #добавляем в список введенное слово в нижнем регистре

"""
В этом блоке мы считаем количество строк на проверку и сами эти строки, а также запишем в виде множества все неправильные слова
"""
task32 = [] #сюда будем складывать строки пословно
task33 = set() #сюда будем складывать неправильные слова

L = int(input()) # задаем количество строк на проверку

for i in range(L):
    task32.append([j for j in input().lower().split()]) #разбиваем введенную строку по пробелу и приводим слова к нижнему регистру
    
    for k in task32[i]: #пробегаем все слова из только что введенной строки
        if k not in task31: #если слова нет в словаре разрешенных слов
            task33.add(k) #добавляем его во множество неправильных слов

#print(task32)
#print(task33)
for i in task33:
    print(i)
```

```python
#Задача 4 - где окажется черепашка
"""
Первая строка ввода - число n, количество команд
Следующие n строк ввода - указания вида "направление расстояние"
На вывод нужны два числа в одной строке - "координата_х координата_y"
"""

task41 = [0, 0]
n = int(input()) #вводим количество команд

for i in range(n): #пробегаем значения от 0 до n-1
    task42 = [j for j in input().lower().split()] #в переменную складываем команду в виде списка [направление, расстояние]
    
    """
    В этом блоке мы выполним проверку на направление и в зависимости от него изменим координату
    """
    if task42[0] == "север":
        task41[1] += int(task42[1]) #прибавляем ко второй координате расстояние
        
    elif task42[0] == "восток":
        task41[0] += int(task42[1]) #прибавляем к первой координате расстояние
        
    elif task42[0] == "юг":
        task41[1] -= int(task42[1]) #отнимаем от второй координаты расстояние
        
    elif task42[0] == "запад":
        task41[0] -= int(task42[1]) #отнимаем от первой координаты расстояние
        
print(str(task41[0]) + " " + str(task41[1])) 
```

```python
#задача 5 - средний рост по классу

"""
В этом блоке считываем все данные в двумерный список
"""
task51 = []
with open("3tasks_input5.tsv") as inf:
    for line in inf:
        task51.append([word for word in line.strip().split()])
#print(task51)

"""
В этом блоке создаем словарь с ключами от 1 до 11 и заполняем его значениями
"""
task52 = {}
for i in range(1, 12): #пробегаем значения от 1 до 11
    task52[i] = []
    
for i in range(len(task51)): #пробегаем все значения от 0 до len(task51)-1
    task52[int(task51[i][0])].append(int(task51[i][2])) #в словарь в список значений по ключу добавляем значение роста
#print(task52)

"""
В этом блоке определим функцию "среднее арифметическое" для списка
"""
def mean(x):
    summ = 0
    ind = 0
    for i in range(len(x)):
        summ += x[i]
        ind += 1
    return summ/ind

"""
В этом блоке выводим средние значения сразу на печать в требуемом формате
"""
for key, value in task52.items():
    if value: #если в значении лежит непустой список
        print(str(key) + " " + str(mean(value))) #выводим ключ и среднее значение списка значений
    else: #иначе
        print(str(key) + " -") #выводим ключ и прочерк
```
